## Perform analysis and visualization

### Subtask:
Based on the data exploration and relationships, perform relevant analyses and create visualizations to understand what the files are conveying. This might include analyzing ad performance, user behavior, campaign effectiveness, etc.


**Reasoning**:
Merge the dataframes based on the identified common columns to create a comprehensive dataset for analysis.



```python
# Merge ad_events_df and ads_df on ad_id
merged_ad_campaign_df = pd.merge(ad_events_df, ads_df, on='ad_id', how='left')

# Merge the result with campaigns_df on campaign_id
merged_ad_campaign_df = pd.merge(merged_ad_campaign_df, campaigns_df, on='campaign_id', how='left')

# Merge the result with users_df on user_id
all_data_df = pd.merge(merged_ad_campaign_df, users_df, on='user_id', how='left')

# Display the first few rows of the combined DataFrame
display(all_data_df.head())
```

**Reasoning**:
Analyze ad performance, campaign effectiveness, and user behavior by grouping the merged dataframe and calculating relevant metrics and summaries. Then, create visualizations to present these findings.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analyze ad performance
ad_performance = all_data_df.groupby('ad_id')['event_type'].value_counts().unstack(fill_value=0)
ad_performance['Total Events'] = ad_performance.sum(axis=1)
ad_performance['CTR'] = (ad_performance['Click'] / ad_performance['Total Events']) * 100
ad_performance['Conversion Rate'] = (ad_performance[['Click', 'Like', 'Share']].sum(axis=1) / ad_performance['Total Events']) * 100

print("Ad Performance:")
display(ad_performance.head())

# Analyze campaign effectiveness
campaign_effectiveness = all_data_df.groupby('campaign_id')['event_type'].value_counts().unstack(fill_value=0)
campaign_effectiveness['Total Events'] = campaign_effectiveness.sum(axis=1)
campaign_effectiveness = pd.merge(campaign_effectiveness, campaigns_df[['campaign_id', 'total_budget', 'name']], on='campaign_id', how='left') # Include 'name' column
campaign_effectiveness['CTR'] = (campaign_effectiveness['Click'] / campaign_effectiveness['Total Events']) * 100
campaign_effectiveness['Conversion Rate'] = (campaign_effectiveness[['Click', 'Like', 'Share']].sum(axis=1) / campaign_effectiveness['Total Events']) * 100
campaign_effectiveness['Cost Per Click'] = campaign_effectiveness['total_budget'] / campaign_effectiveness['Click']
campaign_effectiveness['Cost Per Conversion'] = campaign_effectiveness['total_budget'] / campaign_effectiveness[['Click', 'Like', 'Share']].sum(axis=1)

print("\nCampaign Effectiveness:")
display(campaign_effectiveness.head())


# Analyze user behavior (summary of event counts per user)
user_behavior = all_data_df.groupby('user_id')['event_type'].value_counts().unstack(fill_value=0)
user_behavior['Total Events'] = user_behavior.sum(axis=1)
user_behavior = pd.merge(user_behavior, users_df, on='user_id', how='left')


print("\nUser Behavior (Event Counts per User):")
display(user_behavior.head())


# Visualize Ad Event Distribution
plt.figure(figsize=(8, 6))
all_data_df['event_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel')) # Added color palette
plt.title('Distribution of Ad Event Types')
plt.ylabel('')
plt.show()

# Visualize Top 10 Ads by Total Events
plt.figure(figsize=(12, 6))
ad_performance['Total Events'].sort_values(ascending=False).head(10).plot(kind='bar', color=sns.color_palette('viridis', 10)) # Added color palette
plt.title('Top 10 Ads by Total Events')
plt.xlabel('Ad ID')
plt.ylabel('Total Events')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize Top 10 Campaigns by Conversion Rate
plt.figure(figsize=(12, 6))
campaign_effectiveness['Conversion Rate'].sort_values(ascending=False).head(10).plot(kind='bar', color=sns.color_palette('magma', 10)) # Changed color palette
plt.title('Top 10 Campaigns by Conversion Rate')
plt.xlabel('Campaign ID')
plt.ylabel('Conversion Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize Event Counts by Age Group
plt.figure(figsize=(10, 6))
sns.countplot(data=all_data_df, x='age_group', hue='event_type', palette='tab10') # Changed color palette
plt.title('Ad Event Counts by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Events')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Summarize the key outcomes and insights from the data analysis and visualizations.



```python
# Summarize key findings
print("Key Outcomes and Insights:\n")

# 1. Ad Event Distribution
event_distribution = all_data_df['event_type'].value_counts(normalize=True) * 100
print("1. Ad Event Type Distribution:")
for event, percentage in event_distribution.items():
    print(f"- {event}: {percentage:.2f}%")

# 2. Top Performing Ads (by Total Events and Conversion Rate)
print("\n2. Top Performing Ads:")
top_10_ads_total_events = ad_performance.nlargest(10, 'Total Events')
print("\nTop 10 Ads by Total Events:")
display(top_10_ads_total_events[['Total Events', 'CTR', 'Conversion Rate']])

top_10_ads_conversion_rate = ad_performance.nlargest(10, 'Conversion Rate')
print("\nTop 10 Ads by Conversion Rate:")
display(top_10_ads_conversion_rate[['Total Events', 'CTR', 'Conversion Rate']])


# 3. Top Performing Campaigns (by Conversion Rate and analyze costs)
print("\n3. Top Performing Campaigns:")
top_10_campaigns_conversion_rate = campaign_effectiveness.nlargest(10, 'Conversion Rate')
# Merge campaign names into top_10_campaigns_conversion_rate
top_10_campaigns_conversion_rate = pd.merge(top_10_campaigns_conversion_rate, campaigns_df[['campaign_id', 'name']], on='campaign_id', how='left')

print("\nTop 10 Campaigns by Conversion Rate:")
display(top_10_campaigns_conversion_rate[['Total Events', 'Conversion Rate', 'total_budget', 'Cost Per Click', 'Cost Per Conversion']])

# Analyze campaign costs for top performing campaigns
print("\nCampaign Costs for Top 10 Campaigns by Conversion Rate:")
display(top_10_campaigns_conversion_rate[['name', 'total_budget', 'Cost Per Click', 'Cost Per Conversion']])


# 4. User Behavior by Age Group (from visualization)
print("\n4. User Behavior by Age Group:")
print(" - The countplot shows that 'Impression' is the most frequent event across all age groups.")
print(" - The distribution of other event types (Click, Like, Share, Comment, Purchase) varies by age group, indicating different levels of engagement.")
print(" - Younger age groups (e.g., 18-24, 25-34) appear to have higher counts of interactive events (Click, Like, Share, Comment) compared to older groups, although Impressions are still dominant.")
print(" - Purchase events are relatively low across all age groups.")

# 5. Anomalies or Patterns in Campaign Costs
print("\n5. Anomalies or Patterns in Campaign Costs:")
print(" - Cost per Click and Cost per Conversion vary significantly across campaigns.")
print(" - Some campaigns with high conversion rates might have relatively high or low costs per conversion, suggesting varying efficiency.")
print(" - Further investigation into specific campaigns with high costs per conversion relative to their conversion rates could be beneficial to understand potential inefficiencies.")
```
